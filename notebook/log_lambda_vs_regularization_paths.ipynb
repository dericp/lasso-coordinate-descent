```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


EPSILON = 1e-6


def eval(y, X, w):
    return np.sum(np.square(np.dot(X, w) - y))


def train(lamb, y, X, w):
    # normalization constants
    z = np.square(X).sum(axis=0)

    converged = False
    while not converged:
        converged = True
        for j in range(X.shape[1]):
            j_feature = X[:, j]

            # store the previous value of the jth weight
            prev_w = w[j]

            # set the jth weight to zero---w is now w_j-1
            w[j] = 0

            # get the predictions not considering the jth feature
            predictions = np.dot(X, w)
            residual = y - predictions

            rho = np.dot(j_feature, residual.T)

            if rho < -lamb / 2:
                w[j] = (rho + lamb / 2) / z[j]
            elif rho > lamb / 2:
                w[j] = (rho - lamb / 2) / z[j]
            else:
                w[j] = 0

            if math.fabs(w[j] - prev_w) > EPSILON:
                converged = False

    lamb /= 2

    return w


def get_responses_and_input_variables(df):
    y = df[df.columns[0]]
    X = df.drop(df.columns[0], axis=1).values
    return y, X
```

```python
df_train = pd.read_table('../data/crime-train.txt')
df_test = pd.read_table('../data/crime-test.txt')

y_train, X_train = get_responses_and_input_variables(df_train)
lamb = 600
w_s = list()
w = np.random.normal(0, 1, X_train.shape[1])
log_lambdas = list()
print('training...')
for epoch in range(1, 11):
    print('epoch ' + str(epoch))
    w = train(lamb, y_train, X_train, w)
    w_s.append(np.copy(w))
    log_lambdas.append(math.log (lamb))
    lamb /= 2
```

```python
indices = list()
indices.append(df_train.columns.get_loc('agePct12t29') - 1)
indices.append(df_train.columns.get_loc('pctWSocSec') - 1)
indices.append(df_train.columns.get_loc('PctKids2Par') - 1)
indices.append(df_train.columns.get_loc('PctIlleg') - 1)
indices.append(df_train.columns.get_loc('HousVacant') - 1)
plot_w_s = list()
for w in w_s:
    plot_w_s.append([w[indices[0]], w[indices[1]], w[indices[2]], w[indices[3]], w[indices[4]]])
labels = ['agePct12t29', 'pctWSocSec', 'PctKids2Par', 'PctIlleg', 'HousVacant']
```

```python
for i in range(5):
    plt.plot(log_lambdas, list(x[i] for x in plot_w_s), label=labels[i])
```

```python
plt.legend(bbox_to_anchor=(1.05, 1.05))
plt.xlabel('log(Lambda)')
plt.ylabel('Weight')
plt.title('Weight Regularization Paths')
plt.savefig('../plots/regularization-paths.png', bbox_inches='tight')
plt.show()
```
